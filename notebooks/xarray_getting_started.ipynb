# xarray Getting Started

This notebook computes SPI from a labeled monthly precipitation series and checks that coordinates and metadata survive the xarray adapter path.

```python
import logging

import numpy as np
import pandas as pd
import xarray as xr

from climate_indices import spi
from climate_indices.indices import Distribution

logging.disable(logging.CRITICAL)
```

```python
rng = np.random.default_rng(7)
time = pd.date_range("1980-01-01", periods=40 * 12, freq="MS")
seasonal = 80.0 + 30.0 * np.sin(np.arange(time.size) * 2.0 * np.pi / 12.0)
precip = seasonal + rng.gamma(shape=2.0, scale=5.0, size=time.size)

precip_da = xr.DataArray(
    precip,
    coords={"time": time},
    dims="time",
    name="precipitation",
    attrs={"units": "mm", "long_name": "Monthly precipitation"},
)

spi_6 = spi(
    precip_da,
    scale=6,
    distribution=Distribution.gamma,
    calibration_year_initial=1981,
    calibration_year_final=2010,
)

spi_6
```

```python
assert spi_6.dims == precip_da.dims
assert spi_6.time.identical(precip_da.time)
assert spi_6.attrs["long_name"] == "Standardized Precipitation Index"
assert spi_6.attrs["units"] == "dimensionless"
assert np.isfinite(spi_6.values).sum() > 0

float(spi_6.mean(skipna=True))
```
